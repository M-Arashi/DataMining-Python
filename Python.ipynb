```python
import numpy as np
import pandas as pd

# Set random seed for reproducibility
np.random.seed(42)

# Define parameters for the simulated mixed-effects model
num_samples = 1000  # Number of samples
num_groups = 10  # Number of groups
fixed_effect_coefficients = {'FixedEffect1': 0.5, 'FixedEffect2': -0.3}  # Fixed-effect coefficients
group_effects_std = 0.8  # Standard deviation for random group effects

# Generate synthetic data for the fixed effects
data = pd.DataFrame({
    'FixedEffect1': np.random.randn(num_samples),
    'FixedEffect2': np.random.randn(num_samples),
    'GroupColumn': np.random.randint(0, num_groups, num_samples)  # Simulated group column
})

# Generate random group effects
group_effects = np.random.normal(0, group_effects_std, num_groups)

# Assign random group effects to the data
data['RandomEffect'] = [group_effects[group] for group in data['GroupColumn']]

# Calculate the outcome variable based on the fixed effects and random effects
data['Outcome'] = (fixed_effect_coefficients['FixedEffect1'] * data['FixedEffect1'] +
                   fixed_effect_coefficients['FixedEffect2'] * data['FixedEffect2'] +
                   data['RandomEffect'] + np.random.randn(num_samples))

# Show the simulated data
print(data.head())
```
