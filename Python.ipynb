**Statistical functions**

--- | ---

```python
from numpy.lib.function_base import percentile
x_s = np.array([56.7, 25.5, 50, 58, 55.63, 78.25, 99.5, 15])

print(np.amin(x_s), np.amax(x_s))

print(np.ptp(x_s)) #---> #[np.amax(x_s)-np.amin(x_s)]=np.ptp(x_s)

print(percentile(x_s,50))

print(np.mean(x_s))

print(np.median(x_s))

print(np.std(x_s))

print(np.var(x_s))

print(np.average(x_s , weights=(1, 2, 4, .05, .05, .001, 1 , .4)))
```

```python
import numpy as np
import pandas as pd

# Set random seed for reproducibility
np.random.seed(42)

# Define parameters for the simulated mixed-effects model
num_samples = 1000  # Number of samples
num_groups = 10  # Number of groups
fixed_effect_coefficients = {'FixedEffect1': 0.5, 'FixedEffect2': -0.3}  # Fixed-effect coefficients
group_effects_std = 0.8  # Standard deviation for random group effects

# Generate synthetic data for the fixed effects
data = pd.DataFrame({
    'FixedEffect1': np.random.randn(num_samples),
    'FixedEffect2': np.random.randn(num_samples),
    'GroupColumn': np.random.randint(0, num_groups, num_samples)  # Simulated group column
})

# Generate random group effects
group_effects = np.random.normal(0, group_effects_std, num_groups)

# Assign random group effects to the data
data['RandomEffect'] = [group_effects[group] for group in data['GroupColumn']]

# Calculate the outcome variable based on the fixed effects and random effects
data['Outcome'] = (fixed_effect_coefficients['FixedEffect1'] * data['FixedEffect1'] +
                   fixed_effect_coefficients['FixedEffect2'] * data['FixedEffect2'] +
                   data['RandomEffect'] + np.random.randn(num_samples))

# Show the simulated data
print(data.head())
```

```python
import statsmodels.formula.api as smf
```

```python
# Random Intercept Model
# model allows for different clusters (a group) to have different intercepts
model = smf.mixedlm("weight ~ litsize + C(treatment) + C(sex, Treatment('Male')) + C(treatment):C(sex, Treatment('Male'))",
                    df_3,
                    groups= "litter").fit()

model.summary()
```

```python
# The intercation between treatment and sex is non-significant, therefore this will be removed from the model and re-ran.
# ----- Random Intercept Model w/out Interaction Term
model = smf.mixedlm("weight ~ litsize + C(treatment) + C(sex, Treatment('Male'))", df_3, groups= "litter").fit()
model.summary()
```

```python
# Random Slope Model: No Correlation Between Random Intercept And Slopes
# ----- Random Slope Model: Random intercepts and slopes are independent
model2 = smf.mixedlm("weight ~ litsize + C(treatment) + C(sex)", df_3, groups= "litter",
                     vc_formula = {"sex" : "0 + C(sex)"}).fit()
model2.summary()
```

```python
# ----- Random Slope Model: Random intercepts and slopes are correlated
model3 = smf.mixedlm("weight ~ litsize + C(treatment) + C(sex)", df_3, groups= "litter",
                     re_formula = "1 + C(sex)").fit()
model3.summary()
```

Model Diagnostics
---
The first model will be used which was a random intercept model with the clustering variable being litter.

```python
# Normality
import matplotlib.pyplot as plt
fig = plt.figure(figsize = (16, 9))
import seaborn as sns
ax = sns.distplot(model.resid, hist = False, kde_kws = {"shade" : True, "lw": 1}, fit = stats.norm)

ax.set_title("KDE Plot of Model Residuals (Blue) and Normal Distribution (Black)")
ax.set_xlabel("Residuals")
```

```python
# Q-Q PLot
fig = plt.figure(figsize = (16, 9))
ax = fig.add_subplot(111)

sm.qqplot(model.resid, dist = stats.norm, line = 's', ax = ax)

ax.set_title("Q-Q Plot")
```

There is some deviation from normality, but it doesn't look concerning. There is an outlier, which is evident in both of the plots.

```python
# Shapir-Wilk test of normality
labels = ["Statistic", "p-value"]

norm_res = stats.shapiro(model.resid)

for key, val in dict(zip(labels, norm_res)).items():
    print(key, val)
```

The test is significant which indicates that the assumption of normality for the residuals is violated.

```python
# HOMOSKEDASTICITY OF VARIANCE
fig = plt.figure(figsize = (16, 9))

ax = sns.scatterplot(y = model.resid, x = model.fittedvalues)

ax.set_title("RVF Plot")
ax.set_xlabel("Fitted Values")
ax.set_ylabel("Residuals")
```

```python
# White’s Lagrange Multiplier Test for Heteroscedasticity
from statsmodels.stats.diagnostic import het_white

het_white_res = het_white(model.resid, model.model.exog)

labels = ["LM Statistic", "LM-Test p-value", "F-Statistic", "F-Test p-value"]

for key, val in dict(zip(labels, het_white_res)).items():
    print(key, val)
```

Based on the visual tests, it appears that there isn't much concern about violating the assumption of homoskedasticity of the variance; however the formal testing indicates this assumption is violated.
