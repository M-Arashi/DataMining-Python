**Statistical functions**

Function	| Description
--- | ---
amin, amax	| returns minimum or maximum of an array or along an axis
ptp |	returns range of values (maximum-minimum) of an array or along an axis
percentile(a, p, axis)	| calculate pth percentile of array or along specified axis
median	| compute median of data along specified axis
mean	| compute mean of data along specified axis
std |	compute standard deviation of data along specified axis
var |	compute variance of data along specified axis
average |	compute average of data along specified axis

```python
from numpy.lib.function_base import percentile
x_s = np.array([56.7, 25.5, 50, 58, 55.63, 78.25, 99.5, 15])

print(np.amin(x_s), np.amax(x_s))

print(np.ptp(x_s)) #---> #[np.amax(x_s)-np.amin(x_s)]=np.ptp(x_s)

print(percentile(x_s,50))

print(np.mean(x_s))

print(np.median(x_s))

print(np.std(x_s))

print(np.var(x_s))

print(np.average(x_s , weights=(1, 2, 4, .05, .05, .001, 1 , .4)))
```

```python
A = [[11, 42, 5, 12],
    [-52, 82, 9, 0],
    [-6, 73, 11, 19]]

B = [[1, 49, 51,],
    [-56, 8, 97],
    [-6, 73, 11]]

print(A[0][1]) #The first row and the second column  ---> A[row][column]

print("Rank of A:", np.linalg.matrix_rank(A))

print("Trace of A:", np.trace(A))

print("Rank of B:", np.linalg.matrix_rank(B))

print("Determinant of A:", np.linalg.det(B))

print("Inverse of A:\n", np.linalg.inv(B))
```

```python
import matplotlib.pyplot as plt
labels = ['Category A', 'Category B', 'Category C']
sizes = [30, 45, 35]
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=['lightcoral', 'lightblue', 'lightgreen'])
plt.axis('equal')
plt.title('Pie chart')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

categories = ['Category A', 'Category B', 'Category C']
values1 = [10, 15, 25]
values2 = [5, 20, 15]
values3 = [15, 10, 20]

fig, ax = plt.subplots()

ax.bar(categories, values1, label='Group 1')
ax.bar(categories, values2, bottom=values1, label='Group 2')
ax.bar(categories, values3, bottom=np.array(values1) + np.array(values2), label='Group 3')

ax.set_ylabel('Values')
ax.set_title('Stacked Bar Chart')
ax.legend()

plt.show()
```

```python
import plotly.graph_objects as go


categories = ['Category A', 'Category B', 'Category C']
values1 = [10, 15, 25]
values2 = [5, 20, 15]
values3 = [15, 10, 20]

fig = go.Figure()

fig.add_trace(go.Bar(x=categories, y=values1, name='Group 1'))
fig.add_trace(go.Bar(x=categories, y=values2, name='Group 2'))
fig.add_trace(go.Bar(x=categories, y=values3, name='Group 3'))

fig.update_layout(barmode='stack', title='Stacked Bar Chart')
fig.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = [1, 2, 3, 4, 5]
y1 = [10, 15, 25, 20, 30]
y2 = [5, 10, 15, 10, 20]
y3 = [15, 5, 10, 25, 15]

fig, ax = plt.subplots()

ax.stackplot(x, y1, y2, y3, labels=['Group 1', 'Group 2', 'Group 3'])
ax.set_ylabel('Values')
ax.set_title('Stacked Area Chart')
ax.legend()

plt.show()
```
